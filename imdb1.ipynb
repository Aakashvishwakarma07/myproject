```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.imdb.com/chart/toptv/?ref_=nv_td_tv250"
res=requests.get(url)
s=BeautifulSoup(res.content,'html.parser')
print(s.prettify())
```

```python
a=s.find('td',attrs={'class':'titleColumn'})
print(a.text)
```

```python
print(a.span)
print(a.a)
```

```python
title=[]
rank=[]
crew=[]
year=[]
```

```python
for i in s.find_all('td',attrs={'class':'titleColumn'}):
    title.append(i.a.text)
    rank.append(i.text.strip()[:3])
    crew.append(i.a.get('title'))
    year.append(i.span.text)
```

```python
print(len(title))
print(len(rank))
print(len(crew))
print(len(year))
```

```python
rating=[]
audience=[]
for i in s.find_all('td',attrs={'class':'ratingColumn imdbRating'}):
    rating.append(i.strong.text)
    audience.append(i.strong.get('title')[13:22])
```

```python
print(len(rating))
print(len(audience))
```

```python
imdb=pd.DataFrame({'Title':title,'Rank':rank,'Crew':crew,'Year':year,'Rating':rating,'Audience':audience})
imdb
```

```python
imdb['Rank']=imdb['Rank'].str.replace('.\n','',regex=True)
imdb.head()
```

```python
imdb['Year']=imdb['Year'].str.replace('(','').str.replace(')','')
imdb.head()
```

```python
imdb['Audience']=imdb['Audience'].str.replace(' us','').str.replace(' u','')
imdb.head()
```

```python
imdb['Audience']=imdb['Audience'].str.replace(',','')
imdb.head(50)
```

```python
imdb.to_excel('imdbseries.xlsx',index=False)
```
